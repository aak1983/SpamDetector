```python
import pandas as pd
import plotly_express as px
import wordcloud
import nltk
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/spam_kaggle.csv', encoding='latin-1')
df.head()
```

```python
df.drop(columns=df.iloc[:,2:], axis=1, inplace=True)
df.head()
```

```python
df.rename(columns = {'v1':'label', 'v2':'message'}, inplace=True)
df.head()
```

```python
fig = px.histogram(df, x="label", color="label", color_discrete_sequence=["#871fff","#ffa78c"])
fig.show()
```

```python
fig = px.pie(df.label.value_counts(),labels='index', values='label', color="label", color_discrete_sequence=["#871fff","#ffa78c"] )
fig.show()
```

```python
df['length'] = df['message'].apply(len)
df.head()
```

```python
fig = px.histogram(df, x='length', color='label', color_discrete_sequence=['#871fff','#ffa78c'])
fig.show()                                                                    
```

```python
data_ham = df[df['label'] == 'ham'].copy()
data_spam = df[df['label'] == 'spam'].copy()

def show_word_cloud(df, title):
    text = ' '.join(df['message'].astype(str).tolist())
    stopwords = set(wordcloud.STOPWORDS)
    fig_wordcloud = wordcloud.WordCloud(stopwords=stopwords, background_color='#ffa78c', width = 3000, height = 2000).generate(text)
    plt.figure(figsize=(15,15), frameon=True)
    plt.imshow(fig_wordcloud)
    plt.axis('off')
    plt.title(title, fontsize=20)
    plt.show()
    
show_word_cloud(data_spam, 'Spam Messages')
show_word_cloud(data_ham, 'Ham Messages')
```

```python
df['label'] = df['label'].map({'spam': 1, 'ham':0})
```

```python
# Replace email address with '##emailaddress##'
df['message'] = df['message'].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$', '##emailaddress##')
```

```python
# Replace urls with '##webaddress##'
df['message'] = df['message'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$', '##webaddress##')
```

```python
# Replace money symbol with '##moneysymbol##'
df['message'] = df['message'].str.replace(r'£|\$', '##moneysymbol##')
```

```python
# Replace 10 digit phone number with '##phonenumber##'
df['message'] = df['message'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$', '##phonenumber##')
```

```python
#Replace normal number with '##number##'
df['message'] = df['message'].str.replace(r'\d+(\.\d+)?' ,'##number##')
```

```python
# Remove punctuations
df['message'] = df['message'].str.replace(r'[^\w\d\s]' , ' ')
```

```python
# Remove whitespaces between terms with single space
df['message'] = df['message'].str.replace(r'\s+' , ' ')
```

```python
# Remove leading and trailing whitespaces
df['message'] = df['message'].str.replace(r'^\s+|\s*?$' , ' ')
```

```python
# Modify words to lowercase
df['message'] = df['message'].str.lower()
```

```python
# Import stopwords
import nltk
nltk.download('stopwords')
```

```python
# Remove stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
df['message'] = df['message'].apply(lambda x: ' '.join(term for term in x.split() if term not in stop_words))
```

```python
# Stemming words
ss = nltk.SnowballStemmer('english')
df['message'] = df['message'].apply(lambda x: ' '.join(ss.stem(term) for term in x.split()))
```

```python
# Vectorzing words

sms_df = df['message']
from nltk.tokenize import word_tokenize

# creating a bag-of-words model
all_words = []
for sms in sms_df:
    words = word_tokenize(sms)
    for w in words:
        all_words.append(w)
    
all_words = nltk.FreqDist(all_words)
        
```

```python
print('Number of words: %d' % (len(all_words)))
```

```python
all_words.plot(10, title='Top 10 Most Common Words in Corpus')
```

```python
# Create TfIdf Model
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_model = TfidfVectorizer()
tfidf_vec = tfidf_model.fit_transform(sms_df)


# Serializing the model to a file called model.pkl
import pickle
pickle.dump(tfidf_model, open('model/tfidf_model.pkl','wb'))
tfidf_data = pd.DataFrame(tfidf_vec.toarray())

tfidf_data.head()
```

```python
# Train-Test Splitting
train_size = int(len(tfidf_data) * 0.7)
df_train = tfidf_data.iloc[:train_size]
df_test = tfidf_data.iloc[train_size:]
target = df['label']
df_train['label'] = target

Y = df_train['label']
X = df_train.drop('label', axis=1)

# Splitting training data into train and validation using sklearn
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
import lightgbm as lgb
from sklearn.metrics import f1_score

def train_and_test(model, model_name):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print('F1 Score is: %.6f' % (f1_score(pred, y_test)))
    
for depth in range(1,11):
    lgbmodel = lgb.LGBMClassifier(max_depth=depth, n_estimators=200, num_leaves=40)
    print(f'Max Depth {depth}')
    train_and_test(lgbmodel, "Light GBM")
    print('\n')
```

```python
from sklearn.model_selection import RandomizedSearchCV
lgbmodel_best = lgb.LGBMClassifier(max_depth=7, n_estimators=200, num_leaves=40)
param_grid = {
    'num_leaves': list(range(8, 92, 4)),
    'min_data_in_leaf': [10, 20, 40, 60, 100],
    'max_depth': [3, 4, 5, 6, 8, 12, 16, -1],
    'learning_rate': [0.1, 0.05, 0.01, 0.005],
    'bagging_freq': [3, 4, 5, 6, 7],
    'bagging_fraction': np.linspace(0.6, 0.95, 10),
    'reg_alpha': np.linspace(0.1, 0.95, 10),
    'reg_lambda': np.linspace(0.1, 0.95, 10),
    'min_split_gain': [0.0, 0.1, 0.01],
    'min_child_weight': [0.001, 0.01, 0.1, 0.001],
    'min_child_samples': [20, 30, 25],
    'subsample': [1.0, 0.5, 0.8],
}

model = RandomizedSearchCV(lgbmodel_best, param_grid, random_state=1)
search = model.fit(X_train, y_train)
search.best_params_
```
